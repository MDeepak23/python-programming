````python
sns.heatmap(data.select_dtypes(include=np.number).corr(), cmap='viridis')
````

#OBSERVATION

This heatmap shows the correlation between different variables related to actors and their performance metrics, such as movie count, rating sum, normalized ranks, and Google hits. High correlations (in yellow) indicate strong positive relationships, while low correlations (in purple) suggest weak or negative relationships.

````python
sns.barplot(x='actorName',y='ratingSum',data=d)
````

#OBSERVATION

This bar chart shows the sum of ratings for each actor. Preity Zinta has the highest rating sum, followed by Mithun Chakraborty, while Manoj Joshi has the lowest. This indicates variations in their cumulative ratings.
